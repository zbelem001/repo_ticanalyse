```python
import pandas as pd
from scipy.stats import chi2_contingency as stata
```

```python
df = pd.read_csv('Titanic.xls')
```

```python
df
```

```python
tab = pd.crosstab(df['Pclass'], df['Survived'])
```

```python
tab
```

```python
chi2, p, dof, expected = stata(tab)
```

```python
print("Tableau de contingence :\n", tab)
print(f"\nStatistique du chi² : {chi2:.4f}")
print(f"p-value : {p}")
print(f"Degrés de liberté : {dof}")
print("Effectifs attendus :\n", expected)

# Conclusion
if p < 0.05:
    print("\n✅ Il y a une différence significative de survie entre les sexes.")
else:
    print("\n❌ Aucune différence significative détectée.")
```
